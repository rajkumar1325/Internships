```python
# loadign the dataset
train_dataset = '/root/.cache/kagglehub/datasets/elmadafri/the-wildfire-dataset/versions/3/the_wildfire_dataset_2n_version/train'
validation_directory = '/root/.cache/kagglehub/datasets/elmadafri/the-wildfire-dataset/versions/3/the_wildfire_dataset_2n_version/val'
test_dir = '/root/.cache/kagglehub/datasets/elmadafri/the-wildfire-dataset/versions/3/the_wildfire_dataset_2n_version/test'
```

```python
# List all the classes/directory inside the directory train_dataset
classes = os.listdir(train_dataset)

# to checking how many different tyoes of clasees are present
number_of_classes = len(classes)
print("Number of classes:", number_of_classes)
print("classes present are ", classes)
```

```python
# Visuallsing 1st class
plt.figure(figsize=(15,10))
for i in range(5):
  class_path = os.path.join(train_dataset, classes[1])
  image_name = os.listdir(class_path)[i]
  image_paths = os.path.join(class_path,image_name)
  img = plt.imread(image_paths)

  plt.subplot(1,5,i+1)  #till 1 to 5 images
  plt.imshow(img)
  plt.title(f"{classes[1]} \n shape {img.shape}")
  plt.axis('off')
plt.show
```
